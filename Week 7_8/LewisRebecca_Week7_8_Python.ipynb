```python
#Rebecca Lewis
#Python Scatterplot, Bubble Chart and Density Plot

import pandas as pd

crime = pd.read_csv('C:/Users/rvick/OneDrive/Bellevue/DSC 640/Week 7_8/ex4-2/crimerates-by-state-2005.csv')
```

```python
crime.head()
```

```python
crime = crime[crime['state'] != 'United States']
crime = crime[crime['state'] != 'District of Columbia']
```

```python
crime.reset_index(drop=True, inplace=True)
```

```python
crime
```

```python
from matplotlib import pyplot as plt

```

```python
#Scatter Plot

x= crime['burglary']
y=crime['aggravated_assault']
```

```python
plt.figure(figsize=(20,10))

plt.scatter(x,y)
plt.title('Correlation between Burglaries and Aggravated Assault in the United States')
plt.xlabel('Number of Burglaries')
plt.ylabel('Number of Aggravated Assaults')
plt.xticks(range(0, 1300, 100))
plt.yticks(range(0, 600, 100))
plt.show()
```

```python
#bubble chart
import math

#size down the population so the circle size is based on area
area_pop = crime['population'] / math.pi
area_pop = area_pop.apply(lambda x: math.sqrt(x))
```

```python
plt.figure(figsize=(20,10))

plt.scatter(x,y,s=area_pop, alpha=0.5, label=crime['state'])
plt.title('Correlation between Burglaries and Aggravated Assault in the United States')
plt.xlabel('Number of Burglaries')
plt.ylabel('Number of Aggravated Assaults')
plt.xticks(range(0, 1300, 100))
plt.yticks(range(0, 800, 100))

for i, txt in enumerate(crime['state']):
    plt.annotate(txt, (x[i], y[i]))

plt.show()
```
