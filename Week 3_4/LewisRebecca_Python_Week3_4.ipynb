```python
from matplotlib import pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
population = pd.read_excel('ex2-2\world-population.xlsm')
```

```python
population.head()
```

```python
fig = plt.figure(figsize=(15,8))


plt.plot(population['Year'], population['Population'])
plt.axis([1959, 2010, 2000000000, 8000000000])
plt.xticks(range(1960, 2011, 5))
plt.xlabel('Year')
plt.ylabel('Population (in billions)')
plt.title('Growth of World Population: 1960-2009')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))


plt.step(population['Year'], population['Population'])
plt.axis([1959, 2010, 2000000000, 8000000000])
plt.xticks(range(1960, 2011, 5))
plt.xlabel('Year')
plt.ylabel('Population (in billions)')
plt.title('Growth of World Population: 1960-2009')
plt.show()
```
